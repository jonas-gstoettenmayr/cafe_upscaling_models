## Data Preprocessed

```python
def unlong(data):
    drinks = ["Matcha", "Bubble tea", "Tea", "Hot chocolate", "Espresso", "Latte macchiato", "Cappuccino", "Chai latte", "Pumpkin spice latte", "Iced coffee"]

    subset = data.loc[data["unique_id"] == drinks[0], ["ds"]]

    data2 = pd.DataFrame({"Date": subset["ds"].tolist()})

    for i in range(len(drinks)):
            
        subset = data.loc[data["unique_id"] == drinks[i], ["y"]]

        y = subset["y"].tolist()

        data2[f"{drinks[i]}"] = y

    return data2
```

```python
data_prep = pd.read_parquet("../data/processed_data/df_outliers_capped.parquet") # possibly chnage directory here
data_prep2 = unlong(data_prep)
```

## Engineered features

```python
features = pd.read_parquet("../data/processed_data/fourier+trend+holidays_test.parquet") # possibly chnage directory here
```

```python
fourier_cols = [x for x in features.columns if (('sin' in x) or ('cos' in x)) and ('holiday' not in x)]
fourier_cols
```

```python
def plot_drink_features(df, drink):
    d = df[df["unique_id"] == drink].copy()

    colours = {"Matcha": '#9CB349', 
                "Bubble tea": '#9f85bb', 
                "Tea": '#f0c03f',
                "Hot chocolate": "#8a3d25", 
                "Espresso": "#47231e", 
                "Latte macchiato": "#AD704D", 
                "Cappuccino": "#8b541d", 
                "Chai latte": "#eab287", 
                "Pumpkin spice latte": "#ed6313",
                "Iced coffee": "#7ba4d1"}

    fig = go.Figure()

    fig.add_trace(go.Scatter(
        x=d["ds"],
        y=d["y"],
        mode="lines+markers",
        name="Observed",
        line=dict(color=colours[drink], width=2),
        marker=dict(size=4)
    ))

    from sklearn.linear_model import LinearRegression

    model = LinearRegression()
    model.fit(df[fourier_cols], df['y'])

    df['fourier_combined'] = model.predict(df[fourier_cols])

    fig.add_trace(go.Scatter(
        x=d["ds"],
        y=df['fourier_combined'],
        name="Weekly Seasonality (Fourier)",
        line=dict(color="#5c76e1", width=2, dash="solid")
    ))

    # holidays = d[d["is_holiday"]]

    # for _, row in holidays.iterrows():
    #     fig.add_vline(
    #         x=row["ds"],
    #         line_width=3,
    #         line_dash="solid",
    #         line_color="#39b9ef",
    #         name = "Holiday",
    #         opacity=0.8
    #     )

    fig.update_layout(
        title=dict(
            text=f"{drink} + Engineered Features",
            font=dict(size=26),
            x=0.5
        ),
        template="plotly_dark",
        width=1100,
        height=650,
        margin=dict(t=100),
        plot_bgcolor="#303438",
        paper_bgcolor="#303438",
        xaxis=dict(title="Date", showgrid=True, gridcolor="dimgrey"),
        yaxis=dict(title="Count", showgrid=True, gridcolor='dimgrey'),
        legend=dict(
            orientation="h",
            yanchor="bottom",
            y=1.02,
            xanchor="center",
            x=0.5
        )
    )

    fig.show()
```

```python
plot_drink_features(features, "Tea")
```

## Comparing Best to Baseline and Truth

```python
preds = pd.read_parquet("../data/predictions/final_model_test_set_eval.parquet") # possibly chnage directory here
preds["unique_id"] = preds["unique_id"].apply(lambda x: x.replace('"', ''))
preds
```

```python
def unlong2(data):
    drinks = ["Matcha", "Bubble tea", "Tea", "Hot chocolate", "Espresso", "Latte macchiato", "Cappuccino", "Chai latte", "Pumpkin spice latte", "Iced coffee"]
    # models = data['model_name'].unique().tolist()

    subset = data.loc[(data["unique_id"] == drinks[0]), ["ds"]]

    data2 = pd.DataFrame({"Date": subset["ds"].tolist()})

    for i in range(len(drinks)):
            
        subset = data.loc[(data["unique_id"] == drinks[i]), ["y", "NHITS"]]

        #y = subset["y"].tolist()

        pred = subset["NHITS"].tolist()

        #data2[f"True {drinks[i]}"] = y

        data2[f"PBase {drinks[i]}"] = pred

    for i in range(len(drinks)):
            
        subset = data.loc[(data["unique_id"] == drinks[i]), ["y", "SeasonalNaive"]]

        pred = subset["SeasonalNaive"].tolist()

        data2[f"PBest {drinks[i]}"] = pred

    return data2
```

```python
preds_ = unlong2(preds)
preds_
```

```python
truth_cut = data_prep2[(data_prep2["Date"] >= date.fromisoformat("2025-11-01")) & (data_prep2["Date"] <= date.fromisoformat("2025-12-31"))]
```

```python
truth_cut
```

```python
def plot_drink(drink, df, plot_baseline: bool = True):

    fig = go.Figure()

    colours = {"Matcha": '#9CB349', 
                "Bubble tea": '#9f85bb', 
                "Tea": '#f0c03f',
                "Hot chocolate": "#8a3d25", 
                "Espresso": "#47231e", 
                "Latte macchiato": "#AD704D", 
                "Cappuccino": "#8b541d", 
                "Chai latte": "#eab287", 
                "Pumpkin spice latte": "#ed6313",
                "Iced coffee": "#7ba4d1"}

    
    if plot_baseline:
        fig.add_trace(go.Scatter(
            x=df["Date"],
            y=df[f"PBase {drink}"],
            mode='lines+markers',
            name='Baseline Prediction',
            line=dict(
                color="#408c5c",
                dash='solid',
                width=2
            ),
            marker=dict(size=4)
        ))

    fig.add_trace(go.Scatter(
        x=df["Date"],
        y=df[f"PBest {drink}"],
        mode='lines+markers',
        name='Best Prediction',
        line=dict(
            color="#6e6dc9",
            dash='solid',
            width=2
        ),
        marker=dict(size=4)
    ))

    fig.add_trace(go.Scatter(
        x=truth_cut["Date"],
        y=truth_cut[f"{drink}"],
        mode='lines+markers',
        name='Truth',
        line=dict(
            color=colours[f"{drink}"],
            dash='solid',
            width=2
        ),
        marker=dict(size=4)
    ))

    fig.update_layout(
        title=dict(
            text=f"{drink}",
            font=dict(size=28),
            x=0.5,
            y=0.96
        ),
        xaxis_title="Date",
        yaxis_title="Count",
        template="plotly_dark",
        xaxis=dict(showgrid=True, gridcolor='dimgrey'),
        yaxis=dict(showgrid=True, gridcolor='dimgrey'),
        plot_bgcolor='#303438',
        paper_bgcolor='#303438',
        width=1000, height=600,
        margin=dict(t=100)
    )

    fig.show()
```

## Comparing Errors

```python
def plot_daily_total_error(truth, preds):
    truth["Date"] = pd.to_datetime(truth["Date"])
    preds["Date"] = pd.to_datetime(preds["Date"])

    truth_nov = truth[
        (truth["Date"] >= "2025-11-01") &
        (truth["Date"] <= "2025-11-30")
    ].copy()

    preds_nov = preds[
        (preds["Date"] >= "2025-11-01") &
        (preds["Date"] <= "2025-11-30")
    ].copy()


    df = truth_nov.merge(preds_nov, on="Date", how="inner")

    drinks = truth.columns.drop("Date")

    base_errors = []
    best_errors = []

    for drink in drinks:
        base_errors.append(df[drink] - df[f"PBase {drink}"])
        best_errors.append(df[drink] - df[f"PBest {drink}"])

    df["total_error_base"] = pd.concat(base_errors, axis=1).sum(axis=1)
    df["total_error_best"] = pd.concat(best_errors, axis=1).sum(axis=1)


    fig = go.Figure()

    fig.add_trace(go.Bar(
        x=df["Date"],
        y=df["total_error_base"],
        name="Baseline (Truth - PBase)",
        marker_color="#f0c03f"
    ))

    fig.add_trace(go.Bar(
        x=df["Date"],
        y=df["total_error_best"],
        name="Best (Truth - PBest)",
        marker_color="#6a56f1"
    ))

    fig.add_hline(
        y=0,
        line_width=1,
        line_dash="dash",
        line_color="white",
        opacity=0.6
    )

    fig.update_layout(
        title=dict(
            text="Daily Total Error Across All Drinks (November)",
            font=dict(size=26),
            x=0.5
        ),
        template="plotly_dark",
        barmode="group",
        width=1100,
        height=600,
        margin=dict(t=100),
        plot_bgcolor="#303438",
        paper_bgcolor="#303438",
        xaxis=dict(
            title="Date",
            tickangle=-45,
            showgrid=True, gridcolor='dimgrey'
        ),
        yaxis=dict(title="Total Error (Truth - Prediction)", showgrid=True, gridcolor='dimgrey'),
        legend=dict(
            orientation="h",
            y=1.02,
            x=0.5,
            xanchor="center"
        )
    )

    fig.show()
```

```python
plot_daily_total_error(truth_cut, preds_)
```
